# Examine data quality issues

```python

# Examine data quality issues
print("=" * 80)
print("DATA QUALITY ASSESSMENT")
print("=" * 80)

print("\n1. DUPLICATE DOIs")
print("-" * 80)
# Check for duplicates in DOI
doi_duplicates = relevant_df[relevant_df['doi'].notna()]['doi'].duplicated().sum()
print(f"Duplicate DOIs: {doi_duplicates}")

if doi_duplicates > 0:
    dup_dois = relevant_df[relevant_df.duplicated('doi', keep=False)].sort_values('doi')[['doi', 'title', 'year', 'source']]
    print(f"\nSample duplicates:")
    print(dup_dois.head(10))

print("\n2. DUPLICATE TITLES")
print("-" * 80)
title_duplicates = relevant_df['title'].duplicated().sum()
print(f"Duplicate Titles: {title_duplicates}")

if title_duplicates > 0:
    dup_titles = relevant_df[relevant_df.duplicated('title', keep=False)].sort_values('title')[['title', 'year', 'source', 'doi']]
    print(f"\nSample duplicates (first 5):")
    print(dup_titles.head(10))

print("\n3. YEAR VALIDATION")
print("-" * 80)
invalid_years = relevant_df[(relevant_df['year'] < 2020) | (relevant_df['year'] > 2026)]
print(f"Papers with years outside 2020-2026 range: {len(invalid_years)}")
if len(invalid_years) > 0:
    print(invalid_years[['title', 'year', 'source']])

print("\n4. AUTHOR DATA QUALITY")
print("-" * 80)
no_authors = relevant_df[relevant_df['author_count'] == 0]
print(f"Papers with no authors: {len(no_authors)}")
print(f"Papers with missing author names: {relevant_df['authors'].isna().sum()}")

print("\n5. ABSTRACT AVAILABILITY BY SOURCE")
print("-" * 80)
abstract_by_source = relevant_df.groupby('source').agg({
    'abstract': lambda x: (x.notna() & (x != '')).sum()
}).rename(columns={'abstract': 'with_abstract'})
abstract_by_source['total'] = relevant_df.groupby('source').size()
abstract_by_source['pct'] = (abstract_by_source['with_abstract'] / abstract_by_source['total'] * 100).round(1)
print(abstract_by_source)
```

# Create visualizations to summarize the dataset

```python
# Create visualizations to summarize the dataset
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Set style
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (10, 12)

fig, axes = plt.subplots(4, 1, figsize=(10, 14))
fig.suptitle('Bibliometric Dataset Overview: Survival Analysis & Causality in Demand Forecasting (2020-2026)', 
             fontsize=14, fontweight='bold', y=0.995)

# 1. Publication trends over time
ax1 = axes[0]
year_counts = relevant_df['year'].value_counts().sort_index()
ax1.bar(year_counts.index, year_counts.values, color='steelblue', alpha=0.8, edgecolor='black')
ax1.set_xlabel('Year', fontsize=11, fontweight='bold')
ax1.set_ylabel('Number of Papers', fontsize=11, fontweight='bold')
ax1.set_title('A. Publication Trends (Relevant Papers)', fontsize=12, fontweight='bold', loc='left')
ax1.grid(axis='y', alpha=0.3)
# Add value labels on bars
for i, v in enumerate(year_counts.values):
    ax1.text(year_counts.index[i], v + 5, str(v), ha='center', va='bottom', fontweight='bold')

# 2. Distribution by source
ax2 = axes[1]
source_counts = relevant_df['source'].value_counts()
colors = ['#1f77b4', '#ff7f0e', '#2ca02c']
bars = ax2.barh(source_counts.index, source_counts.values, color=colors, alpha=0.8, edgecolor='black')
ax2.set_xlabel('Number of Papers', fontsize=11, fontweight='bold')
ax2.set_ylabel('Data Source', fontsize=11, fontweight='bold')
ax2.set_title('B. Papers by Data Source', fontsize=12, fontweight='bold', loc='left')
ax2.grid(axis='x', alpha=0.3)
# Add value labels
for i, (bar, v) in enumerate(zip(bars, source_counts.values)):
    ax2.text(v + 10, i, str(v), va='center', fontweight='bold')

# 3. Top 10 research topics
ax3 = axes[2]
topic_counts = topic_freq.head(10)
# Shorten labels for better display
topic_labels = topic_counts['topic_flag'].str.replace('survival:', 'S: ', regex=False).str.replace('forecasting:', 'F: ', regex=False).str.replace('causal:', 'C: ', regex=False)
ax3.barh(range(len(topic_labels)), topic_counts['frequency'], color='mediumseagreen', alpha=0.8, edgecolor='black')
ax3.set_yticks(range(len(topic_labels)))
ax3.set_yticklabels(topic_labels, fontsize=9)
ax3.set_xlabel('Frequency', fontsize=11, fontweight='bold')
ax3.set_ylabel('Research Topic', fontsize=11, fontweight='bold')
ax3.set_title('C. Top 10 Research Topics (S=Survival, F=Forecasting, C=Causal)', fontsize=12, fontweight='bold', loc='left')
ax3.grid(axis='x', alpha=0.3)
ax3.invert_yaxis()
# Add value labels
for i, v in enumerate(topic_counts['frequency']):
    ax3.text(v + 2, i, str(v), va='center', fontweight='bold', fontsize=9)

# 4. Relevance score distribution
ax4 = axes[3]
score_dist = relevant_df['relevance_score'].value_counts().sort_index()
ax4.bar(score_dist.index, score_dist.values, color='coral', alpha=0.8, edgecolor='black', width=0.8)
ax4.set_xlabel('Relevance Score', fontsize=11, fontweight='bold')
ax4.set_ylabel('Number of Papers', fontsize=11, fontweight='bold')
ax4.set_title('D. Distribution of Relevance Scores', fontsize=12, fontweight='bold', loc='left')
ax4.grid(axis='y', alpha=0.3)
ax4.set_xlim(0, 15)

plt.tight_layout()
plt.savefig('dataset_overview.png', dpi=300, bbox_inches='tight')
print("Figure saved as 'dataset_overview.png'")
plt.show()

print("\nVisualization complete!")
```
